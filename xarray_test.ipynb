```python
#Indexing variables in a dataset
da = xr.DataArray(
    np.random.rand(4, 4),
    [
        ("time", pd.date_range("2000-01-01", periods=4)),
        ("space", ["IA", "IL", "IN", "MD"]),
    ],
) 
ds = da.to_dataset(name="Micah was here!") 
ds.isel(space=[1], time=[1]) 
ds.drop_sel(space=['IN', 'IL'])
ds.drop_dims('time')  
```

```python
# Reindexing 
da = xr.DataArray(
    np.random.rand(4, 4),
    [
        ("time", pd.date_range("2000-01-01", periods=4)),
        ("space", ["IA", "IL", "IN", "MD"]),
    ],
)   
da.reindex(space = ['IA', 'CA']) 
mwh = da.rename('Micah was here!')
foo = (10 * mwh[:2, :2]).rename('foo')
```

```python
# Masking 
da = xr.DataArray(np.arange(16).reshape(4, 4), dims=["x", "y"])
print(da) 
da.where(da.x + da.y < 4) 

print() # Interesting that it won't show both arrays
da.where(da.y < 2, drop = True)
```
